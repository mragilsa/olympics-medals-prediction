```python
teams = pd.read_csv("teams.csv")
```

```python
teams = teams[["team", "country", "year", "athletes", "age", "prev_medals", "medals"]]
```

```python
teams = teams.dropna()
```

```python
teams
```

```python
train = teams[teams["year"] < 2012].copy()
test = teams[teams["year"] >= 2012].copy()
```

```python
train.shape
```

```python
test.shape
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
```

```python
predictors = ["athletes", "prev_medals"]
target = "medals"
```

```python
reg.fit(train[predictors], train[target])
```

```python
predictions = reg.predict(test[predictors])
```

```python
test["predictions"] = predictions
```

```python
test
```

```python
test.loc[test["predictions"] < 0, "predictions"] = 0
```

```python
test["predictions"] = test["predictions"].round()
```

```python
test
```

```python
import matplotlib.pyplot as plt

test_sorted = test.sort_values("medals")

plt.figure(figsize=(10, 6))

plt.plot(test_sorted["medals"], test_sorted["predictions"], color="red", label="Prediction Line")

plt.xlabel("Actual Medals")
plt.ylabel("Predicted Medals")
plt.title("Line Chart: Actual vs Predicted")
plt.legend()
plt.grid(True)

plt.show()
```

```python
from sklearn.metrics import mean_absolute_error

error = mean_absolute_error(test["medals"], test["predictions"])
```
